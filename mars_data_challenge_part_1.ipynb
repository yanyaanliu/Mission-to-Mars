```python
# Imports
import json
import pandas as pd


from splinter import Browser
from bs4 import BeautifulSoup as soup
from webdriver_manager.chrome import ChromeDriverManager

# Set up Browser with Splinter
executable_path = {'executable_path': ChromeDriverManager().install()}
browser = Browser('chrome', **executable_path, headless=False)
```

```python
# Make the browser navigate to the website and get the content
url = 'https://redplanetscience.com/'
browser.visit(url)
html = browser.html
soup = soup(html, 'html.parser')
```

```python
# Preview
soup
```

```python
#Scrapping title for every article
titles=soup.find_all('div',class_='content_title')

i=1
for title in titles:
    print(i,".",title.text)
    i=i+1
    
```

```python
#Scrapping the Summary text for each article
summaries= soup.find_all('div', class_='article_teaser_body')

b=1
for summary in summaries:
    print(b,'.',summary.text)
    b=b+1
```

```python
element = soup.find_all('div',class_='list_text')

# Iterate through all the elements to scrap the titles and the summary for each article
mars_list = []
for elem in element:
    mars_data={"title":"","preview":""}
    title=elem.find('div',class_='content_title').text
    summary= elem.find('div', class_='article_teaser_body').text
    mars_data["title"] = title
    mars_data["preview"] = summary
    mars_list.append(mars_data)
print(mars_list)
```

```python
# Export list to JSON file
json_mars = json.dumps(mars_list)
print(json_mars)
```

```python
# Save json as csv file
df=pd.read_json(json_mars)
df.to_csv('marsdata.csv',index=False)
```
